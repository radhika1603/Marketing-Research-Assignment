```python
from googleapiclient.discovery import build
from textblob import TextBlob

# API Key
api_key = "AIzaSyBWE45pJvUtYRWV-qV8Q503ATelN-POBcI"
video_id = "tYDKo3lUqDo"

# Initialize YouTube API client
youtube = build('youtube', 'v3', developerKey=api_key)

def get_video_comments(video_id, max_comments=100):
    comments = []
    next_page_token = None

    while len(comments) < max_comments:
        request = youtube.commentThreads().list(
            part="snippet",
            videoId=video_id,
            maxResults=50,
            pageToken=next_page_token,
            textFormat="plainText"
        )
        response = request.execute()

        for item in response.get("items", []):
            comment = item["snippet"]["topLevelComment"]["snippet"]["textDisplay"]
            comments.append(comment)
            if len(comments) >= max_comments:
                break

        next_page_token = response.get("nextPageToken")
        if not next_page_token:
            break

    return comments

# Fetch comments
comments = get_video_comments(video_id, max_comments=100)

# Analyze sentiment
for i, comment in enumerate(comments, start=1):
    sentiment_score = TextBlob(comment).sentiment.polarity
    sentiment = "Positive" if sentiment_score > 0 else "Negative" if sentiment_score < 0 else "Neutral"
    print(f"{i}. {comment}\n   Sentiment: {sentiment} (Score: {sentiment_score})\n")
```

```python
import matplotlib.pyplot as plt
from googleapiclient.discovery import build
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk

# Download VADER lexicon
nltk.download('vader_lexicon')

# API Key
api_key = "AIzaSyBWE45pJvUtYRWV-qV8Q503ATelN-POBcI"
video_id = "tYDKo3lUqDo"

# Initialize YouTube API client
youtube = build('youtube', 'v3', developerKey=api_key)

# Function to fetch comments
def get_video_comments(video_id, max_comments=100):
    comments = []
    next_page_token = None

    while len(comments) < max_comments:
        request = youtube.commentThreads().list(
            part="snippet",
            videoId=video_id,
            maxResults=50,
            pageToken=next_page_token,
            textFormat="plainText"
        )
        response = request.execute()

        for item in response.get("items", []):
            comment = item["snippet"]["topLevelComment"]["snippet"]["textDisplay"]
            comments.append(comment)
            if len(comments) >= max_comments:
                break

        next_page_token = response.get("nextPageToken")
        if not next_page_token:
            break

    return comments

# Fetch comments
comments = get_video_comments(video_id, max_comments=100)

# Initialize VADER for sentiment analysis
sia = SentimentIntensityAnalyzer()

# Sentiment classification
sentiment_counts = {"Positive": 0, "Neutral": 0, "Negative": 0}

for comment in comments:
    sentiment_score = sia.polarity_scores(comment)['compound']
    if sentiment_score > 0.05:
        sentiment_counts["Positive"] += 1
    elif sentiment_score < -0.05:
        sentiment_counts["Negative"] += 1
    else:
        sentiment_counts["Neutral"] += 1

# Plot sentiment distribution
plt.figure(figsize=(8, 5))
plt.bar(sentiment_counts.keys(), sentiment_counts.values(), color=['green', 'gray', 'red'])
plt.xlabel("Sentiment")
plt.ylabel("Number of Comments")
plt.title("YouTube Comments Sentiment Analysis")
plt.show()
```

```python
from prettytable import PrettyTable
from googleapiclient.discovery import build
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk

# Download VADER lexicon
nltk.download('vader_lexicon')

# API Key
api_key = "AIzaSyBWE45pJvUtYRWV-qV8Q503ATelN-POBcI"
video_id = "tYDKo3lUqDo"

# Initialize YouTube API client
youtube = build('youtube', 'v3', developerKey=api_key)

# Function to fetch comments
def get_video_comments(video_id, max_comments=100):
    comments = []
    next_page_token = None

    while len(comments) < max_comments:
        request = youtube.commentThreads().list(
            part="snippet",
            videoId=video_id,
            maxResults=50,
            pageToken=next_page_token,
            textFormat="plainText"
        )
        response = request.execute()

        for item in response.get("items", []):
            comment = item["snippet"]["topLevelComment"]["snippet"]["textDisplay"]
            comments.append(comment)
            if len(comments) >= max_comments:
                break

        next_page_token = response.get("nextPageToken")
        if not next_page_token:
            break

    return comments

# Fetch comments
comments = get_video_comments(video_id, max_comments=100)

# Initialize VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Create a PrettyTable for displaying comments
table = PrettyTable()
table.field_names = ["Index", "Comment", "Sentiment", "Score"]

# Analyze sentiment and add rows to the table
for i, comment in enumerate(comments, start=1):
    sentiment_score = sia.polarity_scores(comment)['compound']
    sentiment = "Positive" if sentiment_score > 0.05 else "Negative" if sentiment_score < -0.05 else "Neutral"
    table.add_row([i, comment[:50] + ("..." if len(comment) > 50 else ""), sentiment, round(sentiment_score, 2)])

# Print table
print(table)
```

```python
import pandas as pd
df = pd.DataFrame(table._rows, columns=table.field_names)
df.to_csv("youtube_comments.csv", index=False)
```
